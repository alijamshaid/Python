```python
days=[1,2,3,4,5,6,7]
max_t=[34,65,43,67,12,33,1]
min_t=[23,23,32,34,43,6,75]
avg_t=[34,56,34,76,21,39,5]
plt.title("nowshera")
#plt.xlabel("days")
plt.plot(days,max_t,label="max_t")
plt.plot(days,min_t,label="min_t")
plt.plot(days,avg_t,label="avg_t")
plt.legend(loc="best",shadow=True,fontsize="large")
plt.grid()
```

```python
import numpy as np
company=["GOOGLE","FB","AMAZON","TWITTER"]
revenue=[34,65,45,74]
profit=[22,34,28,55]
ypos=np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
#plt.ylabel("revenue (billions)")
plt.title("silicpon valley buisness")
plt.bar(ypos-0.2,revenue,width=0.4,label="revenue")
plt.bar(ypos+0.2,profit,width=0.4,label="profit")
plt.legend()
```

```python
plt.yticks(ypos,company)
#plt.ylabel("revenue (billions)")
plt.title("silicpon valley buisness")
plt.barh(ypos-0.2,revenue,label="revenue")
plt.barh(ypos+0.2,profit,label="profit")
plt.legend()
```

```python
blood_test_men=[23,43,35,76,45,67,67,83,22,34,65]
blood_test_females=[29,34,53,67,12,42,76,38,12,43,56]
plt.hist([blood_test_men,blood_test_females],bins=[10,30,50,100],rwidth=0.95,color=['green','red'],
        label=['men','women'],orientation="horizontal")
plt.legend()
plt.title("blood test")
```
